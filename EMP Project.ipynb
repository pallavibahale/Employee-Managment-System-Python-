```python
import mysql.connector
from tkinter import *  # we importing all from te tkinter
from tkinter import messagebox
from tkinter import ttk
win=Tk()
win.title('EMPLOYEE MANAGEMENT SYSTEM')
win.minsize(width=800,height=600)
win.configure(bg='skyblue')


def show():
     
    com=mysql.connector.connect(user='root',password='pallavi',host='localhost' , database='project1')
    qur='select* from emp'
    mycur=com.cursor()
    mycur.execute(qur)
    data=mycur.fetchall() # fetch all the date in one column # return list into tup
    for i in data:
        treev.insert("",'end',value=(i[0],i[1],i[3],i[2]))  # end para
    mycur.close()
    com.close()
    
    
    
def add():

    name=n1.get()  # acces value using variable
    mobno=m1.get()
    dept=d1.get()
    emailid=e1.get()
    # all value should compulsory
    if name=='' or mobno=='' or dept=='' or emailid=='':
        messagebox.showinfo('info','all fild are compulsory')
    else:
        import mysql.connector
        com=mysql.connector.connect(user='root',password='pallavi',host='localhost'
        , database='project1')
        qur='INSERT INTO emp(name,mobno,dept,emailid) VALUES("%s","%s","%s","%s")'%(name,mobno,dept,emailid)
        mycur=com.cursor()
        mycur.execute(qur)
        com.commit() # if you have to modify something then use it
        mycur.close()
        com.close()
        # messagebox
        messagebox.showinfo('info','data is added successfully')
        n1.set('')
        m1.set('')
        d1.set('')
        e1.set('')                                                                                                                                                                      
        

def salary():
    root=Tk()
    root.title('SALARY')
    root.minsize(width=1200,height=1000)
    root.configure(bg='cyan')
    def disp(): # defind disp fun can defind outside as well
        name=e9.get() # entry name e9 gloabl variable
        com=mysql.connector.connect(user='root',password='pallavi',host='localhost'
        , database='project1')
        qur='SELECT mobno from emp where name = "%s"'%(name) # select and show it
        mycur=com.cursor()
        mycur.execute(qur)
        sal=mycur.fetchone()
        l10.config(text=sal[0]+" Rs")
        mycur.close()
        com.close()
        
    l9=Label(root,text='Name of Employee',
           bd=5,relief='ridge',bg='white',
          fg='black',width=20,font=('times new roman',12,'bold'))

    l9.place(x=190,y=80)
    # entry
    var3=StringVar() # defind variable
    e9=Entry(root,bd=5,textvariable=var3,relief='ridge',bg='white',
          fg='black',width=40,font=('times new roman',12,'bold'))
    e9.place(x=385,y=80)
    # button

    b9=Button(root,text='Show Salary',width=20,command=disp)
    b9.place(x=395,y=120)
    # create label
    l10=Label(root,text='Salary is here',height=4)
    l10.place(x=190,y=160)
    root.mainloop()
def Delete():
    name=var1.get()
    if  name=='' :
        messagebox.showinfo('info','all fild are compulsory')
    else:
        com=mysql.connector.connect(user='root',password='pallavi',host='localhost'
        , database='project1')
        qur='DELETE from emp WHERE name="%s"'%(name)
        mycur=com.cursor()
        mycur.execute(qur)
        com.commit() # if you have to modify something then use it
        mycur.close()
        com.close()
        # messagebox
    messagebox.showinfo('info','data is deleted successfully')
    var1.set('')
    
def Select():
    # fetch date from treeview to this column
    name=var2.get()
    
    if  name=='' :
        messagebox.showinfo('info','all fild are compulsory')
    else:
        com=mysql.connector.connect(user='root',password='pallavi',host='localhost'
        , database='project1')
        qur='select*from emp where name="%s"'%(name)
        mycur=com.cursor()
        mycur.execute(qur)
        list1=mycur.fetchall() # show list becz its all value
        for i in list1:
            en.insert(0,i[0])
            e2.insert(0,i[1])
            e3.insert(0,i[2])
            e4.insert(0,i[3])
        mycur.close()
        com.close()
        
        var2.set('')
    
def Update():
    # we are going to get nnall four condition
    
    name=n1.get()  # acces value using variable
    mobno=m1.get()
    dept=d1.get()
    emailid=e1.get()
    # all value should be compulsory
    if name=='' or mobno=='' or dept=='' or emailid=='':
        messagebox.showinfo('info','all fild are compulsory')
    else:
        import mysql.connector
        com=mysql.connector.connect(user='root',password='pallavi',host='localhost'
        , database='project1')
        qur= 'UPDATE emp SET mobno="%s",dept="%s",emailid="%s" where name="%s"'%(mobno,dept,emailid,name)
        mycur=com.cursor()
        mycur.execute(qur)
        com.commit() # if you have to modify something then use it
        mycur.close()
        com.close()
        # messagebox
        messagebox.showinfo('info','data is added successfully')
        n1.set('')
        m1.set('')
        d1.set('')
        e1.set('')
def Clear():
    treev.delete(*treev.get_children())   # get.children built fun it will clrea all column of the table * all vlue
n1=StringVar()
m1=StringVar()
d1=StringVar()
e1=StringVar()
# main label
lbl=Label(win,text='EMPLOYEE MANAGEMENT SYSTEM',
           bd=5,relief='ridge',bg='white',
          fg='black',width=40,font=('times new roman',18,'bold'))

lbl.place(x=268,y=40)

# name label and entry
l1=Label(win,text='NAME',
           bd=5,relief='ridge',bg='white',
          fg='black',width=20,font=('times new roman',12,'bold'))

l1.place(x=190,y=100)

en=Entry(win,bd=5,textvariable=n1,relief='ridge',bg='white',
          fg='black',width=40,font=('times new roman',12,'bold'))
en.place(x=385,y=100)

# mobno label and entry
l2=Label(win,text='MOBILE NUMBER',
           bd=5,relief='ridge',bg='white',
          fg='black',width=20,font=('times new roman',12,'bold'))

l2.place(x=190,y=140)

e2=Entry(win,bd=5,textvariable=m1,relief='ridge',bg='white',
          fg='black',width=40,font=('times new roman',12,'bold'))
e2.place(x=385,y=140)

# dept label and entry
l3=Label(win,text='Department',
           bd=5,relief='ridge',bg='white',
          fg='black',width=20,font=('times new roman',12,'bold'))

l3.place(x=190,y=180)

e3=Entry(win,bd=5,textvariable=d1,relief='ridge',bg='white',
          fg='black',width=40,font=('times new roman',12,'bold'))
e3.place(x=385,y=180)

# email id
l4=Label(win,text='Email ID',
           bd=5,relief='ridge',bg='white',
          fg='black',width=20,font=('times new roman',12,'bold'))

l4.place(x=190,y=220)

e4=Entry(win,bd=5,textvariable=e1,relief='ridge',bg='white',
          fg='black',width=40,font=('times new roman',12,'bold'))
e4.place(x=385,y=220)

# buttons
# Add,Exit,Show,salary
showbtn=Button(win,text='Show',width=20,command=show)
showbtn.place(x=200,y=320)
addbtn=Button(win,text='Add',width=20,command=add)
addbtn.place(x=500,y=320)
exitbtn=Button(win,text='Exit',width=20,command=win.destroy)
exitbtn.place(x=200,y=360)
salbtn=Button(win,text='Salary',command=salary,width=20)
salbtn.place(x=500,y=360)
# delete .....label,entry,button
# label
l5=Label(win,text='Delete By Name',
           bd=5,relief='ridge',bg='white',
          fg='black',width=20,font=('times new roman',12,'bold'))

l5.place(x=190,y=400)
# entry
var1=StringVar()
e5=Entry(win,bd=5,textvariable=var1,relief='ridge',bg='white',
          fg='black',width=40,font=('times new roman',12,'bold'))
e5.place(x=385,y=400)
# button

delbutton=Button(win,text='Delete',width=20,command=Delete)
delbutton.place(x=395,y=440)

# update ... label ,entry,select and update button
# label

l6=Label(win,text='Update By Name',
           bd=5,relief='ridge',bg='white',
          fg='black',width=20,font=('times new roman',12,'bold'))

l6.place(x=190,y=480)
var2=StringVar() # variable
# entry
e6=Entry(win,bd=5,textvariable=var2,relief='ridge',bg='white',
          fg='black',width=40,font=('times new roman',12,'bold'))
e6.place(x=385,y=480)
# button2
selbutton=Button(win,text='Select',width=20,command=Select)
selbutton.place(x=200,y=520)
upbutton=Button(win,text='Update',width=20,command=Update)
upbutton.place(x=395,y=520)


# treeview

treev = ttk.Treeview(win, selectmode ='browse',height=20) # treev is obj
treev.place(x=900,y=100,width=450)

treev["columns"] = ("1","2","3","4") # create column

treev['show'] = 'headings'  # show heading
treev.column("1", width = 90, anchor ='c') # center 
treev.column("2", width = 90, anchor ='se') # south east (c nantar)
treev.column("3", width = 90, anchor ='se')
treev.column("4", width = 90, anchor ='se')

treev.heading("1", text ="Name") # puting heading in that column
treev.heading("2", text ="Mobile No")
treev.heading("3", text ="Dept")
treev.heading("4", text ="EmailID")

# clear button
clearbutton=Button(win,text='Clear',width=20,command=Clear)
clearbutton.place(x=1080,y=560)

# add 
win.mainloop() # for open this window
```
